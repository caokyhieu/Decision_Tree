```python
missing_column=['Age','Cabin','Embarked']
```

```python
#drop 'Cabin' feature because missing too much
train_df = train_df.drop(['Cabin'],axis=1)
```

```python
nomial_feature = ['Survived','Pclass','Sex','Ticket','Embarked']
continuous_feature = list(set(train_df.columns) - set(nomial_feature))
```

```python
#filling missing data
# 'Age'
train_df[['Age']] = train_df[['Age']].fillna(train_df['Age'].mean())
```

```python
#'Embarked'
train_df[['Embarked']] = train_df[['Embarked']].fillna(train_df['Age'].mode())
```

```python
import matplotlib.pyplot as plt
grid = sns.FacetGrid(train_df[['SibSp'] +['Sex','Embarked']],row='Embarked' ,hue='Sex')
grid.map(sns.distplot,'SibSp')
grid.add_legend()
plt.show()
```
